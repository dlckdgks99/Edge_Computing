```python
import soundfile as sf
def trim_audio_data(audio_file, save_file, start, end):
    y, sr = librosa.load(audio_file,16000)

    ny = y[sr*start:sr*end]

    sf.write(save_file, ny, sr)
```

```python
import wave
import contextlib
import os
import librosa

path_dir = '/home/leecs/2022KCC/experiment/datagenerate/collectData/clean/' #경로 끝에 / 꼭 붙이기
save_dir = '/home/leecs/2022KCC/experiment/datagenerate/newData/clean/'
file_list =os.listdir(path_dir) #경로 읽어 파일명 리스트 만들기
file_list.sort() #정렬

acc = 0
l = 0
for i in file_list:
    print(path_dir+i)
    with contextlib.closing(wave.open(path_dir + i,'r')) as f:
        frames = f.getnframes()
        rate = f.getframerate()
        duration = frames / float(rate)
        print(duration)
        trim_audio_data(path_dir+i, save_dir+i, 0, int(duration))
        acc += duration #누적값
        l += 1

#결과 출력
print(l)
print (acc)
```
